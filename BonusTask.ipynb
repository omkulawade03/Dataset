```python
print("--- Q1: Loading the Titanic Dataset ---")

titanic_df = pd.read_csv('titanic.csv')
print(titanic_df.head())
print("\n")
```

```python
print("--- Q2: Basic Information & Missing Values ---")
print("Dataset Shape:", titanic_df.shape)
print("\nMissing Values per column:")
print(titanic_df.isnull().sum())
print("\n")
```

```python
print("--- Q3: Survival Count Plot ---")
# Visualizing how many people survived vs died
plt.figure(figsize=(6, 4))
# Using hue and legend=False to prevent seaborn warnings
sns.countplot(data=titanic_df, x='Survived', hue='Survived', palette='pastel', legend=False)
plt.title('Count of Survival (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
print("--- Q4: Survival by Passenger Class (Pclass) ---")
# Visualizing survival based on their ticket class
plt.figure(figsize=(8, 5))
sns.countplot(data=titanic_df, x='Pclass', hue='Survived', palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
print("--- Q5: Survival by Gender ---")
# Visualizing survival based on gender
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Sex', hue='Survived', palette='Set1')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
print("--- Q6: Age Distribution ---")
# Checking the age distribution of passengers
plt.figure(figsize=(8, 5))
sns.histplot(titanic_df['Age'].dropna(), kde=True, bins=30, color='purple')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
