```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import joblib
```

```python
# =====================================================================
# Q1. (Preparing Features)
# =====================================================================
print("--- Q1: Preparing Features ---")
# Load the cleaned dataset
df = pd.read_csv("ford_car_dataset.csv")

# Separate into Independent Features (X) and Dependent Feature (Y)
X = df.drop(columns=['price'])
Y = df['price']

# Print shapes
print(f"Shape of X: {X.shape}")
print(f"Shape of Y: {Y.shape}\n")
```

```python
# =====================================================================
# Q2. (One-Hot Encoding)
# =====================================================================
print("--- Q2: One-Hot Encoding ---")
# Identify all categorical columns in X
categorical_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
print(f"Categorical Columns Identified: {categorical_cols}")

# Perform One-Hot Encoding using pd.get_dummies()
# Using drop_first=True helps prevent multi-collinearity issues in Linear Regression
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Convert all newly created encoded columns to integer type (0 or 1)
encoded_cols = [col for col in X_encoded.columns if col not in X.columns or col in categorical_cols]
X_encoded[encoded_cols] = X_encoded[encoded_cols].astype(int)

print("\nFirst 5 rows after encoding:")
print(X_encoded.head())
print("\n")
```

```python
# =====================================================================
# Q3. (Feature Scaling)
# =====================================================================
print("--- Q3: Feature Scaling ---")
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply Standard Scaling on numerical columns
X_encoded[numerical_cols] = scaler.fit_transform(X_encoded[numerical_cols])

print("First 5 rows of the scaled features:")
print(X_encoded[numerical_cols].head())
print("\n")
```

```python
# =====================================================================
# Q4. (Train-Test Split)
# =====================================================================
print("--- Q4: Train-Test Split ---")
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, Y, test_size=0.33, random_state=42
)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test:  {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test:  {y_test.shape}\n")
```

```python
# =====================================================================
# Q5. (Building Linear Regression Model)
# =====================================================================
print("--- Q5: Building Linear Regression Model ---")
# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Print intercept and coefficients
print(f"Model Intercept: {model.intercept_:.4f}\n")
print("Model Coefficients:")
for col, coef in zip(X_train.columns, model.coef_):
    print(f"  {col}: {coef:.4f}")
print("\n")
```

```python
# =====================================================================
# Q6. (Making Predictions)
# =====================================================================
print("--- Q6: Making Predictions ---")
# Make predictions on the test data
y_pred = model.predict(X_test)

# Compare the first 10 predicted vs actual values
comparison_df = pd.DataFrame({
    'Actual Value': y_test.values[:10],
    'Predicted Value': y_pred[:10]
})
print("First 10 Predicted vs Actual comparison:")
print(comparison_df)
print("\n")
```

```python
# =====================================================================
# Q7. (Model Evaluation using R2 Score)
# =====================================================================
print("--- Q7: Model Evaluation using R2 Score ---")
r2 = r2_score(y_test, y_pred)
print(f"R2 Score: {r2:.4f}")
print("\n*Interpretation written in the section below* \n")
```

```python
# =====================================================================
# Q8. (Saving the Model)
# =====================================================================
print("--- Q8: Saving the Model ---")
# Extract column list to preserve the structure for future deployment
columns_list = X_encoded.columns.tolist()

# Save objects using joblib
joblib.dump(model, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(columns_list, "columns.pkl")

print("Successfully saved 'LR_ford_car.pkl', 'scaler.pkl', and 'columns.pkl'.")
```
