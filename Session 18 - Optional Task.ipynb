```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('amazon.csv')

# Let's check missing values in raw data first
missing_counts_raw = df.isnull().sum()
missing_pct_raw = (df.isnull().sum() / len(df)) * 100

print("--- Raw Missing Values ---")
for col in df.columns:
    if missing_counts_raw[col] > 0:
        print(f"{col}: {missing_counts_raw[col]} missing ({missing_pct_raw[col]:.3f}%)")

# Let's check for the '|' issue in rating
pipe_count = (df['rating'] == '|').sum()
print(f"Number of '|' values in rating: {pipe_count}")

# Check duplicates in raw data
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Now let's apply the correct cleaning to see if any other missing values show up
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)
df['rating'] = pd.to_numeric(df['rating'].astype(str).str.replace('|', '', regex=False), errors='coerce')
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)

# Recalculate missing values after numeric coercion
missing_counts_clean = df.isnull().sum()
missing_pct_clean = (df.isnull().sum() / len(df)) * 100

print("\n--- Missing Values After Numeric Conversion ---")
for col in df.columns:
    if missing_counts_clean[col] > 0:
        print(f"{col}: {missing_counts_clean[col]} missing ({missing_pct_clean[col]:.3f}%)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
df = pd.read_csv('amazon.csv')

# 2. Extract the primary top-level category from the nested hierarchy
df['main_category'] = df['category'].apply(lambda x: str(x).split('|')[0])

# =========================================================
# Q6. Count Plots of Categorical Features
# =========================================================

# Step A: Identify categorical columns in the cleaned dataset
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("Identified Categorical Columns in Dataset:")
print(list(categorical_cols))
print("\n" + "="*60 + "\n")

# Step B: Create the count plot for the main categorical feature
plt.figure(figsize=(12, 6))

# Order the bars by highest frequency count for clean visual hierarchy
category_order = df['main_category'].value_counts().index

sns.countplot(
    data=df,
    y='main_category',
    order=category_order,
    palette='viridis',
    hue='main_category',
    legend=False
)

plt.title('Product Distribution across Main Retail Categories', fontsize=14, pad=15)
plt.xlabel('Number of Product Listings', fontsize=12)
plt.ylabel('Primary Category', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and clean the dataset completely
df = pd.read_csv('amazon.csv')

for col in ['discounted_price', 'actual_price']:
    df[col] = df[col].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').astype(float)
df['rating'] = pd.to_numeric(df['rating'].astype(str).str.replace('|', '', regex=False), errors='coerce')
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').astype(float)

df['rating'] = df['rating'].fillna(df['rating'].median())
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())
df = df.drop_duplicates()

# =========================================================
# Q7. Correlation Heatmap
# =========================================================
numeric_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

# Create the correlation matrix
corr_matrix = df[numeric_cols].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,          # Show correlation values inside the squares
    cmap='coolwarm',     # Red for positive, blue for negative correlation
    fmt=".2f",          # Limit to 2 decimal places
    linewidths=0.5,      # Add sharp lines between squares
    vmin=-1, vmax=1      # Fix scale boundaries from -1 to 1
)

plt.title('Correlation Heatmap of Numeric Features', fontsize=14, pad=15)
plt.tight_layout()
plt.show()
```
